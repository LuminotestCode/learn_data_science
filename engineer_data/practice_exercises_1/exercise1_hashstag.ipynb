# Ejercicio 1 -> Dividir los hashtag

* En el archivo de hashtag , se encuentran una serie de palabras separadas por `#` , en una misma fila hay varias palabras , mediante lo aprendido de los RDD separar las palabras y dejar solamente un vector de palabras, luego contar la cantidad de veces que se repite cada palabra

### 1. Iniciar sesión en spark

```python
#Desde pyspark.sql importamos la SparkSession
import findspark #Librería para encontrar donde está instalado spark
findspark.init() 

from pyspark.sql import SparkSession

#Desde SparkSession comenzamos a construir la conexión
spark = SparkSession.builder \
    .appName("exercise1_hashtag") \
    .master("local[*]") \
    .getOrCreate()

sc = spark.sparkContext
```

```python
#Importamos el archivo txt mediante textFile
path_file = './hashstag_file.txt'
hashtag_file = sc.textFile(path_file, minPartitions = 2)
#hashtag_file.getNumPartitions()
```

```python
#Ahora mediante flatMap  y lambda recorremos el archivo separando por #
array_words = hashtag_file.flatMap(lambda word: word.split('#'))
header = array_words.first()
array_without_spaces = array_words.filter(lambda word : word != '' and word != header)
array_without_spaces.take(15)
```

```python
array_without_spaces.countByValue()
```
