# 🐋 Fundamentos de docker

### 💡 ¿Qué es Docker?

Docker es una aplicación de código abierto, que permite crear espacios ligeros llamados `contenedores` que permiten,
alojar  todas las dependencias, paquetes y configuraciones necesarias para que funcione una aplicación sin necesidad de 
ocupar toda la  capacidad de la infraestructura de un servidor.

### 🧠 2. Conceptos básicos de Docker

| Concepto           | Definición                                                            |
| ------------------ | --------------------------------------------------------------------- |
| **Contenedor**     | Entorno aislado donde se ejecuta una aplicación con sus dependencias. |
| **Imagen (Image)** | Plantilla inmutable que se usa para crear contenedores.               |
| **Dockerfile**     | Script de instrucciones para construir una imagen.                    |
| **Docker Hub**     | Repositorio de imágenes Docker (público por defecto).                 |
| **Volume**         | Método para persistir datos generados y usados por contenedores.      |
| **Network**        | Red virtual que permite la comunicación entre contenedores.           |
| **Bind Mount**     | Vincula un directorio del host con uno del contenedor.                |

### 💡 7. Buenas prácticas

* Usar imágenes oficiales y específicas (ej. python:3.11-slim en lugar de python).

* No guardar secretos en el Dockerfile.

* Usar `.dockerignore` para evitar copiar archivos innecesarios.

* Limpia los contenedores e imágenes que no uses con `docker system prune.`

* Para producción, considera usar imágenes distroless o minimalistas.
