## List comprehensions Python []

List comprehensions se define mediante una lista , es decir , con los corchetes cuadrados y se define un iterable que ser√° el que genere los elementos de la lista

> üí°Primero de la forma cl√°sica sin list comprehensions

```python
numbers  = []
for element in range(1, 12):
    numbers.append(element)
print(numbers)
```

> üí°Ahora mediante list comprehensions en una √∫nica l√≠nea

```python
numbers_v1 = [element for element in range(1,12)]
numbers_v1
```

### Operaciones y condicionales dentro de list comprenhensions

Es posible agregar condiciones dentro de las listas , de manera que los elementos se generen solo si se cumplen ciertos requerimientos

```python
numbers_v2 = [i*2 for i in range(1, 40) if i % 2 == 0 ]
numbers_v2
```

## Dictionary comprehensions {}

la sintaxis es muy similar a las list comprehensions , solo que se debe tener en cuenta que estos se componen de clave valor

```python
dict_1 = {f'value {i}' : i*2 for i in range(1, 10)}
dict_1
```

Tambi√©n podr√≠amos tomar las llaves o los valores de otra estructura de datos (listas, otros diccionarios etc..)

```python
countries = ['col', 'mex', 'bol', 'pe']
population = ['52','130','12','33']  #Millones de personas
dict_countries_population = {countries[i] : population[i] for i in range(0, len(countries))}
dict_countries_population
```

> üí°Hay un manera m√°s √≥ptima de generar este diccionario, debido a que hay una funci√≥n destinada a unir una lista con otra

```python
dict_countries_population_tuple = list(zip(countries, population))
dict_countries_population_tuple
```

```python
dict_countries_population = {countries:population for (countries,population) in zip(countries,population)}
dict_countries_population
```

De igual manera que con las listas , los diccionarios tambi√©n se pueden generar en base a condicionales
